## 1. Importing Required Libraries

In this step, the necessary Python libraries are imported for data manipulation and analysis.

```python
import pandas as pd
import numpy as np
```

## 2. Loading the Dataset

The dataset is loaded into a Pandas DataFrame to begin the data cleaning process.

```python
df = pd.read_excel("Dataset for Data Analytics.xlsx")
print("Dataset Shape:", df.shape)
```

```python
df.head()
```

## 3. Dataset Overview

This section examines the dataset structure, number of records, columns, and data types.

```python
print("Rows and Columns:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
df.info()
```

### Observation

The dataset contains 1200 records and 14 columns. Most columns contain complete data, while the CouponCode column contains missing values that require cleaning.

## 4. Missing Value Analysis

The dataset is checked for null or missing values to identify incomplete information.

```python
df.isnull().sum()
```

### Observation

The CouponCode column contains 309 missing values. All other columns have complete information.

```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

```python
missing_percentage = (
    df.isnull().sum() / len(df)
) * 100

missing_percentage[missing_percentage > 0]
```

## 5. Handling Missing Values

Missing CouponCode values are replaced with "No Coupon" because a blank value indicates that no promotional coupon was used during the purchase.

```python
df["CouponCode"].value_counts(dropna=False)
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
df.isnull().sum()
```

### Observation

All missing values have been successfully handled and the dataset now contains zero null values.

## 6. Duplicate Record Detection

The dataset is checked for duplicate records and duplicate Order IDs to ensure uniqueness and data integrity.

```python
df.duplicated().sum()
```

```python
df["OrderID"].duplicated().sum()
```

### Observation

No duplicate rows and no duplicate Order IDs were found in the dataset.

## 7. Date Format Standardization

Dates are converted into a consistent YYYY-MM-DD format for improved readability and compatibility with analytical tools.

```python
df["Date"].head()
```

```python
df["Date"].dtype
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")
```

```python
df["Date"].head()
```

### Observation

All dates were successfully standardized into a uniform format.

## 8. Text Data Cleaning

Text columns are cleaned by removing unnecessary spaces and standardizing capitalization.

```python
text_columns = [
    "Product",
    "PaymentMethod",
    "OrderStatus",
    "ReferralSource",
    "CouponCode"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()
```

### Observation

Text fields are now consistently formatted, reducing the possibility of duplicate categories caused by inconsistent spelling or capitalization.

## 9. Business Rule Validation

The TotalPrice column is validated by comparing it against Quantity × UnitPrice.

```python
calculated_total = (
    df["Quantity"] *
    df["UnitPrice"]
).round(2)

mismatch = df[
    calculated_total !=
    df["TotalPrice"].round(2)
]

len(mismatch)
```

### Observation

No inconsistencies were found between the calculated and recorded total prices.

## 10. Final Verification

A final quality check is performed to confirm that the dataset is clean and ready for analysis.

```python
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows:", df.duplicated().sum())
print("Duplicate Order IDs:", df["OrderID"].duplicated().sum())
```

### Observation

The final dataset contains:
- 0 Missing Values
- 0 Duplicate Rows
- 0 Duplicate Order IDs

The dataset is fully cleaned and prepared for analysis.

## 11. Exporting the Cleaned Dataset

The cleaned dataset is saved for future analysis and reporting.

```python
df.to_excel("cleaned_dataset.xlsx", index=False)
```
