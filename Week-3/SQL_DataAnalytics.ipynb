## Creating SQL Database

To perform SQL-based analysis, the dataset is converted into a SQLite database table. This allows us to execute SQL queries directly on the data and extract meaningful business insights using filtering, sorting, grouping, and aggregation operations.

```python
import sqlite3
```

```python
# Create SQLite connection
conn = sqlite3.connect('ecommerce.db')

# Convert DataFrame into SQL table
df.to_sql(
    name='orders',
    con=conn,
    if_exists='replace',
    index=False
)

print("Database and table created successfully!")
```

```python
# Verify table creation
query = """
SELECT *
FROM orders
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

### Observation

The dataset has been successfully stored as a SQL table named **orders**. We can now perform SQL queries to analyze sales, customer behavior, payment methods, order status, and revenue trends.

## Basic SQL Queries

Basic SQL queries help us retrieve and inspect data from the database table. The SELECT statement is used to fetch records and specific columns from the dataset.

```python
query = """
SELECT *
FROM orders
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query displays the first 10 records from the orders table, providing an overview of the available data and its structure.

```python
query = """
SELECT OrderID, Product, Quantity, TotalPrice
FROM orders
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

### Observation

Only the relevant columns are displayed, making the output more focused and easier to analyze.

```python
query = """
SELECT DISTINCT Product
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

The DISTINCT keyword removes duplicate values and returns the list of unique products available in the dataset.

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query calculates the total number of orders present in the dataset.

```python
query = """
SELECT COUNT(DISTINCT CustomerID) AS Unique_Customers
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query determines the number of unique customers who placed orders, helping us understand the customer base size.

## Data Filtering using WHERE Clause

The WHERE clause is used to filter records based on specific conditions. It helps extract relevant information from the dataset by selecting only those rows that satisfy the given criteria.

```python
query = """
SELECT *
FROM orders
WHERE OrderStatus = 'Delivered';
"""

pd.read_sql_query(query, conn)
```

### Observation

This query retrieves all orders that were successfully delivered to customers.

```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
WHERE TotalPrice > 5000;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query identifies high-value orders with a total price exceeding ₹5000.

```python
query = """
SELECT OrderID, CustomerID, PaymentMethod, TotalPrice
FROM orders
WHERE PaymentMethod = 'Credit Card';
"""

pd.read_sql_query(query, conn)
```

### Observation

This query filters orders that were paid using the Credit Card payment method.

```python
query = """
SELECT OrderID, Product, Quantity
FROM orders
WHERE Quantity > 3;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query displays orders where customers purchased more than three units of a product.

```python
query = """
SELECT OrderID, CustomerID, ReferralSource
FROM orders
WHERE ReferralSource = 'Social Media';
"""

pd.read_sql_query(query, conn)
```

### Observation

This query identifies customers who reached the platform through social media referrals.

```python
query = """
SELECT OrderID, Product, TotalPrice, OrderStatus
FROM orders
WHERE TotalPrice > 3000
AND OrderStatus = 'Delivered';
"""

pd.read_sql_query(query, conn)
```

### Observation

The query returns delivered orders with a total price greater than ₹3000.

```python
query = """
SELECT OrderID, Product, OrderStatus
FROM orders
WHERE OrderStatus = 'Cancelled'
OR OrderStatus = 'Returned';
"""

pd.read_sql_query(query, conn)
```

### Observation

This query displays orders that were either cancelled or returned.

## Sorting Data using ORDER BY

The ORDER BY clause is used to sort records in ascending (ASC) or descending (DESC) order. It helps identify top-performing products, highest-value orders, and other ranked business metrics.

```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query sorts orders from highest to lowest total price, helping identify the most valuable transactions.

```python
query = """
SELECT OrderID, Product, TotalPrice
FROM orders
ORDER BY TotalPrice ASC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query displays orders starting from the lowest transaction value.

```python
query = """
SELECT DISTINCT Product
FROM orders
ORDER BY Product ASC;
"""

pd.read_sql_query(query, conn)
```

### Observation

Products are arranged in alphabetical order, making them easier to browse and locate.

```python
query = """
SELECT CustomerID, TotalPrice
FROM orders
ORDER BY TotalPrice DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query ranks customers based on their order value, highlighting high-spending customers.

```python
query = """
SELECT OrderID, Product, Quantity
FROM orders
ORDER BY Quantity DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query identifies orders containing the largest quantities of products purchased.

```python
query = """
SELECT OrderID, Date, Product, TotalPrice
FROM orders
ORDER BY Date DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

Orders are sorted from the most recent to the oldest based on the transaction date.

## Aggregate Functions Analysis

Aggregate functions are used to perform calculations on multiple rows of data and return a single summarized result. These functions help in understanding overall business performance, revenue generation, customer activity, and sales trends.

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query calculates the total number of orders present in the dataset.

```python
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query computes the overall revenue generated from all customer orders.

```python
query = """
SELECT AVG(TotalPrice) AS Average_Order_Value
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query determines the average amount spent per order.

```python
query = """
SELECT MAX(TotalPrice) AS Highest_Order_Value
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query identifies the highest-value transaction in the dataset.

```python
query = """
SELECT MIN(TotalPrice) AS Lowest_Order_Value
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query returns the smallest transaction amount recorded in the dataset.

```python
query = """
SELECT SUM(Quantity) AS Total_Quantity_Sold
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query calculates the total number of product units sold across all orders.

```python
query = """
SELECT AVG(Quantity) AS Average_Quantity
FROM orders;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query measures the average number of products purchased in each order.

## Grouping Data using GROUP BY

The GROUP BY clause is used to arrange data into groups based on one or more columns. It is commonly combined with aggregate functions to generate meaningful business insights such as revenue by product, orders by payment method, and customer spending patterns.

```python
query = """
SELECT Product,
       SUM(TotalPrice) AS Total_Revenue
FROM orders
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query calculates revenue generated by each product and ranks them from highest to lowest revenue.

```python
query = """
SELECT PaymentMethod,
       COUNT(*) AS Total_Orders
FROM orders
GROUP BY PaymentMethod
ORDER BY Total_Orders DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query identifies the most preferred payment methods among customers.

```python
query = """
SELECT OrderStatus,
       COUNT(*) AS Total_Orders
FROM orders
GROUP BY OrderStatus
ORDER BY Total_Orders DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query shows the distribution of orders across different statuses such as Delivered, Shipped, Returned, and Cancelled.

```python
query = """
SELECT ReferralSource,
       SUM(TotalPrice) AS Total_Revenue
FROM orders
GROUP BY ReferralSource
ORDER BY Total_Revenue DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query evaluates which referral channels contribute the highest revenue to the business.

```python
query = """
SELECT CustomerID,
       SUM(TotalPrice) AS Total_Spending
FROM orders
GROUP BY CustomerID
ORDER BY Total_Spending DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query identifies the top 10 customers based on their total spending amount.

```python
query = """
SELECT Product,
       SUM(Quantity) AS Quantity_Sold
FROM orders
GROUP BY Product
ORDER BY Quantity_Sold DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query highlights the products with the highest sales volume.

```python
query = """
SELECT PaymentMethod,
       AVG(TotalPrice) AS Average_Order_Value
FROM orders
GROUP BY PaymentMethod
ORDER BY Average_Order_Value DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

This query compares the average spending of customers across different payment methods.

```python
query = """
SELECT ReferralSource,
       AVG(TotalPrice) AS Average_Revenue
FROM orders
GROUP BY ReferralSource
ORDER BY Average_Revenue DESC;
"""

pd.read_sql_query(query, conn)
```

### Observation

The query helps identify which referral channels attract higher-value customers on average.
